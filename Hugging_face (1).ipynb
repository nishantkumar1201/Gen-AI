```python
#Sentiment Analysis
from transformers import pipeline
```

```python
#Load the spam detection pipeline
spam_classifier=pipeline("text-classification",model="philschmid/distilbert-base-multilingual-cased-sentiment")
```

```python
txt=[
    "Congratulations! you have won a 500000 INR Amazon gift card, click here to claim",
    "Hi Nitish Lets meet today at 2PM",
    "URGENT:Your gmail acct has compromised,click here to secure it "
]
results=spam_classifier(txt)
label_mapping={
    "negative":"Negative",
    "neutral":"Neutral",
    "positive":"Positive"
}
for result in results:
    label=result["label"]
    score=result["score"]
    # Use the label directly from the model output as the key
    sentiment=label_mapping[label]
    print(f"Sentiment:{sentiment},Confidence Score:{score:.2f}")
```

**Text Summarization**

```python
from transformers import AutoModelForSeq2SeqLM,AutoTokenizer
```

```python
#Load a pretrained model & tokenizer
model_name="facebook/bart-large-cnn"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
txt=""" Generative AI is a type of artificial intelligence (AI) that creates new, original content, such as text, images, music, video, and code, by learning patterns from existing data.
Unlike traditional AI, which analyzes or categorizes data, generative AI produces unique outputs in response to user prompts. These models are trained on vast datasets to mimic human creativity and are powered by foundation models,
which are large AI models capable of performing complex, multi-tasking functions """
```

```python
#Tokenize the input text
inputs=tokenizer.encode("summarize:"+txt,return_tensors="pt",max_length=1024,truncation=True)
```

```python
#Generate the summary
summary_ids=model.generate(inputs,max_length=150,num_beams=4,early_stopping=True)
summary=tokenizer.decode(summary_ids[0],skip_special_tokens=True)
print("Summary:",summary)
```

**Text Translation**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

translation_model_name = "Helsinki-NLP/opus-mt-en-es"
translation_tokenizer = AutoTokenizer.from_pretrained(translation_model_name)
translation_model = AutoModelForSeq2SeqLM.from_pretrained(translation_model_name)

input_text="My name is Nitish & I like study AI"
```

```python
input_ids=translation_tokenizer.encode(input_text,return_tensors="pt")
```

```python
#Generate the translated text
translated_ids=translation_model.generate(input_ids,max_length=150,num_beams=4,early_stopping=True)
translated_text=translation_tokenizer.decode(translated_ids[0],skip_special_tokens=True)
print("Translated Text:",translated_text)
```
